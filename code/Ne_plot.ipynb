```r
library(ggplot2)

# Create the time points and Ne values
time_points <- c(0, 36, 36, 100)  # Using 100 as an arbitrary end point
ne_values <- c(300, 300, 30000, 30000)

# Set the maximum time for the plot
max_time <- 72
```

```r
par(mfrow = c(2, 1))  # Set up a 1x2 panel for two plots

# Create the first plot
plot(1, type = "n", 
     xlim = c(max_time, 0),  # Reversed x-axis
     ylim = c(0, max(ne_values) * 1.1),
     xlab = "",
     ylab = expression(N[e]),
     main = "Constant",
     cex.lab = 1.3,     # Axis labels size
     cex.main = 1.5,    # Title size
     cex.axis = 1.1)    # axis numbers size

# Draw the blue step lines with transparency
lines(time_points, ne_values, type = "s", lwd = 2, col = rgb(0, 0, 1, 0.5))  # Blue with 50% transparency

# # Draw the red horizontal line
abline(h = 30000, col = rgb(1, 0, 0, 0.5), lwd = 2)  # Red with 50% transparency

abline(v = 36, col = "black", lty = 2, lwd = 1.5)  # Vertical dashed line at x=36
# Add annotation for the crash point
text(36, 27000, "t = 36", pos = 4, cex = 1.5, col = "black")


# Create the fluctuating step lines
time_points_fluctuate <- seq(0, 100, by = 6)
ne_values_red <- ifelse(seq_along(time_points_fluctuate) %% 2 == 0, 300, 30000)
ne_values_blue <- ifelse(time_points_fluctuate < 36, 300, ne_values_red)

# Create the second plot in the panel
plot(1, type = "n", 
    xlim = c(max_time, 0),  # Reversed x-axis
    ylim = c(0, max(ne_values_red) * 1.1),
    xlab = "Generations ago (past -> present)",
    ylab = expression(N[e]),
    main = "Seasonal",
    cex.lab = 1.3,     # Larger axis labels
    cex.main = 1.5,    # Larger title
    cex.axis = 1.1)    # Larger axis numbers

# Draw the blue step line
lines(time_points_fluctuate, ne_values_blue, type = "s", lwd = 2, col = rgb(0, 0, 1, 0.5))
# Draw the red step line
lines(time_points_fluctuate, ne_values_red, type = "s", lwd = 2, col = rgb(1, 0, 0, 0.5))

abline(v = 36, col = "black", lty = 2, lwd = 1.5)
text(36, 31500, "t = 36", pos = 4, cex = 1.5, col = "black")
```
